### Imports

```python
from pycaret.datasets import get_data
from pycaret.regression import *
```

### Load data 

```python
dataset = get_data('insurance')
```

### Experiment 1 on data - using PyCaret setup

```python
first_setup = setup(dataset, target = 'charges', session_id=1997)
```

```python
lr = create_model('lr')
```

```python
plot_model(lr,plot = 'residuals')
```

### Experiment 2 on data - using PyCaret setup

```python
second_setup = setup(dataset, target = 'charges',session_id=1997,
           normalize = True,
           polynomial_features = True, trigonometry_features = True, feature_interaction=True, 
           bin_numeric_features= ['age', 'bmi'])
```

```python
second_setup
```

```python
second_setup[16].columns #checking the new columns/features created by the PyCaret
```

```python
lr = create_model('lr') #R2 score increased significantly
```

```python
plot_model(lr,plot = 'residuals')
```

```python
# save transformation pipeline and model 
save_model(lr, model_name = 'insurance_prediction_model')
```
